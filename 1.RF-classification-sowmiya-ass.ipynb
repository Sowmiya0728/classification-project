```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load datset
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
#Convert categorical variables to dummy variables
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
# split dataset into independent and dependent variables
indep=dataset[["age","bp","al","su","bgr","bu","sc","sod","pot","hrmo","pc_normal","pcc_present","ba_present","htn_yes","dm_yes","cad_yes","appet_yes","pe_yes","ane_yes"]]
```

```python
dep=dataset["classification_yes"]
```

```python
indep.shape
```

```python
dep
```

```python
#split dataset into training and test sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
#initialize and train Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy', random_state=0)
classifier.fit(X_train,y_train)
```

```python
#Evaluate the model
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
```

```python
print(clf_report)
```

```python
#predict using the trained model 
sample_data=[[40,70,1,0,100,30,1,140,4,15,False,False,False,True,True,False,True,False,True]]
prediction=classifier.predict(sample_data)
```
